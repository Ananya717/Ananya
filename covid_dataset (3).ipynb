```python
#Q3.low data level understanding
#a.find count of unique value
unique_values= c['location'].unique()
unique_values
```

```python
value_counts=c['location'].value_counts()
value_counts
```

```python
#b.maximum frequency values count
max_frequency = value_counts.max()
max_frequency
```

```python
#c.find maximum and mean value in 'total cases'.
max_value = c['total_cases'].max()
mean_value = c['total_cases'].mean()
max_value
mean_value
```

```python
#d.find 25%,50%,75% quartile value in 'total_deaths'
import numpy as np
quartiles = np.percentile(c['total_deaths'], [25, 50, 75])
print("Quartile values for 'total deaths':")
print("25%:", quartiles[0])
print("50%:", quartiles[1])
print("75%:", quartiles[2])
```

```python
#e. find which continent maximum 'human_development_index'.
max_hdi_continent = c.loc[c['human_development_index'].idxmax()]['continent']
max_hdi_continent
```

```python
#f. find which continent has minimum 'gdp_per_capita'.
min_gdp_continent = c.loc[c['gdp_per_capita'].idxmin()]['continent']
min_gdp_continent
```

```python
#Q4.filter the dataframe  with only this column
#['continent','location','date','total_cases','total_death','gdp_per_capita','human_development_index']update the dataframe.
df=['continent', 'location', 'date', 'total_cases', 'total_death', 'gdp_per_capita', 'human_development_index']
df=pd.DataFrame(c)
print("Updated DataFrame:")
print(df)
```

```python
#Q5.data cleaning
#a. remove all duplicate options
df_no_duplicates = df.drop_duplicates()
print(df_no_duplicates)
```

```python
#b.find missing values in all columns
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
#c.remove all observation where continent column value is missing.
df_filtered =c.dropna(subset=['continent'])
print("DataFrame after removing missing values in 'continent' column:")
print(df_filtered)
```

```python
#d. fill all missing values with 0
df_filled = df.fillna(0)
print("DataFrame after filling missing values with 0:")
print(df_filled)
```

```python
#Q6.data time format
#a.convert date column in datetime format using pandas.to_datetime
df['date'] = pd.to_datetime(df['date'])
print("DataFrame after converting 'date' column to datetime:")
print(df)
```

```python
#b.create a new column month after extracting month data from date column
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.month
print("DataFrame with the new 'month' column:")
print(df)
```

```python
#Q7.Data Aggregation
#a. Find max value in all columns using groupby function on 'continent' column
max_values_by_continent = df.groupby('continent').max()
print("Maximum values by continent:")
print(max_values_by_continent)
```

```python
#b. store the result in anew dataframe named 'df_groupby'
df_groupby = df.groupby('continent').max().reset_index()
print("Maximum values by continent stored in df_groupby:")
print(df_groupby)
```

```python
#Q8.Feature Engineering
#a.Create a new feature 'total_deaths_to_total_cases'by ratio of 'total_deaths'column to 'total_cases'
df['total_deaths_to_total_cases'] = df['total_deaths'] / df['total_cases']
print("DataFrame with the new feature 'total_deaths_to_total_cases':")
print(df)
```

```python
#Q9 Data visualization
#a.perform univariate analysis on 'gdp_per_capita'column by plotting histogram using seaborn dist plot.
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(10, 6))
sns.distplot(df['gdp_per_capita'], bins=20, kde=False, color='green')

plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.title('Histogram of GDP per Capita')

plt.show()
```

```python
#b.Plot a scatter plot of 'total_cases'& 'gdp_per_capita'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_cases', y='gdp_per_capita', data=df, color='red', marker='o')
plt.xlabel('Total Cases')
plt.ylabel('GDP per Capita')
plt.title('Scatter Plot of Total Cases vs. GDP per Capita')

plt.show()
```
